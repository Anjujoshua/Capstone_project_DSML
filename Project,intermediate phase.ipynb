```python
# Select numerical columns from df2
numerical_cols = df2.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Function to detect outliers using IQR
def detect_outliers_iqr(df2, columns):
    outlier_indices = {}

    for col in columns:
        Q1 = df2[col].quantile(0.25)
        Q3 = df2[col].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers
        outliers = df2[(df2[col] < lower_bound) | (df2[col] > upper_bound)][col]
print(outliers)
```

```python
# Boxplot before outlier removal
numerical_cols = df2.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Plot boxplots to visualize outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=df2[numerical_cols])
plt.xticks(rotation=90) 
plt.title("Boxplot Before Outlier Removal")
plt.show()
```
